```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Created the folder
if not os.path.exists('result/visualizations'):
    os.makedirs('result/visualizations')

# Visualization 1: Showed Distribution of Math Scores
plt.figure(figsize=(6, 4))
plt.hist(df['math score'], bins=10, edgecolor='black', color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.savefig('result/visualizations/math_scores_distribution.png')
plt.close()

# Visualization 2: Showed Average Scores by Gender
gender_avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

gender_avg_scores.plot(kind='bar', figsize=(6, 4), color=['blue', 'green', 'orange'])
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('result/visualizations/average_scores_by_gender.png')
plt.close()

# Visualization 3: Showed Impact of Test Preparation on Math Scores
prep_avg_scores = df.groupby('test preparation course')['math score'].mean()

plt.figure(figsize=(6, 4))
prep_avg_scores.plot(kind='bar', color=['green', 'red'])
plt.title('Impact of Test Preparation on Math Scores')
plt.xlabel('Test Preparation Status')
plt.ylabel('Average Math Score')
plt.tight_layout()
plt.savefig('result/visualizations/test_prep_impact.png')
plt.close()

# Visualization 4: Showed Average Scores by Parental Education Level
parent_education_scores = df.groupby('parental level of education')['math score'].mean()

plt.figure(figsize=(8, 4))
parent_education_scores.plot(kind='bar', color='purple')
plt.title('Average Math Scores by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Math Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('result/visualizations/parental_education_scores.png')
plt.close()

# Visualization 5: Showed Performance Across Different Race/Ethnicity Groups
race_scores = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

race_scores.plot(kind='line', marker='o', figsize=(6, 4))
plt.title('Performance Across Different Race/Ethnicity Groups')
plt.xlabel('Race/Ethnicity Group')
plt.ylabel('Average Scores')
plt.tight_layout()
plt.savefig('result/visualizations/race_ethnicity_performance.png')
plt.close()

print("All visualizations are saved in the 'result/visualizations' folder.")
```

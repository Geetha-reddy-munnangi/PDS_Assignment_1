```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
df
```

```python
# changing the frailty to numerical column
df['Frailty'] = df['Frailty'].map({'N': 0, 'Y': 1})
df
```

```python
# Calculating correlations between everything
correlation_weight_frailty = df['Weight'].corr(df['Frailty'])
correlation_height_frailty = df['Height'].corr(df['Frailty'])
correlation_age_frailty = df['Age'].corr(df['Frailty'])

# Printing the  correlations between each of them
print("Correlation between Weight and Frailty:", correlation_weight_frailty)
print("Correlation between Height and Frailty:", correlation_height_frailty)
print("Correlation between Age and Frailty:", correlation_age_frailty)
```

```python
# Saving the  results to the file
with open('results.txt', 'w') as file:
    file.write("Results of Data Analysis:\n")
    file.write("-" * 30 + "\n")
    file.write(f"Correlation between Weight and Frailty: {correlation_weight_frailty:.4f}\n")
    file.write(f"Correlation between Height and Frailty: {correlation_height_frailty:.4f}\n")
    file.write(f"Correlation between Age and Frailty: {correlation_age_frailty:.4f}\n")
```

```python
# Setting up the graph background
sns.set(style="whitegrid")

# Plotting the  correlation between Weight and Frailty
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.regplot(x='Weight', y='Frailty', data=df, logistic=True, ci=None)
plt.title(f'Weight vs Frailty\nCorrelation: {correlation_weight_frailty:.2f}')


# Plotting the correlation between Height and Frailty
plt.subplot(1, 3, 2)
sns.regplot(x='Height', y='Frailty', data=df, logistic=True, ci=None)
plt.title(f'Height vs Frailty\nCorrelation: {correlation_height_frailty:.2f}')


# Plotting the correlation between Age and Frailty
plt.subplot(1, 3, 3)
sns.regplot(x='Age', y='Frailty', data=df, logistic=True, ci=None)
plt.title(f'Age vs Frailty\nCorrelation: {correlation_age_frailty:.2f}')
plt.savefig('graphs.png')

plt.show()
```
